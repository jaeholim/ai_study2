```python
# 복잡한 복합문
for i in range(5):
    print(i)
else:               # for와 짝을 이루는 else - for를 완벽하게 실행되면 else 구문이 실행 됨
    print("언제나와?")
    
print("==============")
for i in range(5):
    print(i)
    if i == 3:
        break
else:
    print("end")     # for문이 끝까지 실행되었는지 체크 하는데 쓰면 좋음
print("==============")
# 절차적 프로그램 - 문을 쓰는 순서가 중요함
print("==============")

for i in range(5):
    if i == 4:
        continue
    print(i)
else:
    print('end')    # continue문 사용시는 for문이 끝까지 실행된것이기 때문에 else 구문이 실행됨
    
print("==============")
for i in range(2, 10):
    for j in range(1, 10):
        print(i, '*', j, '=', i*j, end=', ')
    else:
        print()
        print(i, '단 완료')
else:
    print('구구단 완료')
    
print("==============")
for i in range(2, 10):        
    for j in range(1, 10):
        if j==5:
            break                             # 바로위의 loop만 break함
        print(i, '*', j, '=', i*j, end=', ')   
    else:
        print()
        print(i, '단 완료')
else:
    print('구구단 완료')

    
print("==============")
for i in range(2, 10):        
    for j in range(1, 10):
        if j==5:
            continue                             # 바로위의 loop만 continue
        print(i, '*', j, '=', i*j, end=', ')   
    else:
        print(i, '단 완료')
else:
    print('구구단 완료')
    
    
print("==============")
for i in range(2, 10): 
    if i%2 == 0:
        continue
    for j in range(1, 10):
        print(i, '*', j, '=', i*j, end=', ')   
    else:
        print(i, '단 완료')
else:
    print('구구단 완료')
```

```python
for i in 1,2,3,:    # 튜플에서 괄호를 생략할 수 있기 때문에 가능
    print(i)

for i in {3,2,1,}:  # set(집합)은 순서가 없지만 내부적으로 오름 차순으로 정렬하여 표현됨 - 사용자가 정의한 순서대로 유지되지 않음
    print(i)
    
for i in '임재호':  
    print(i)
    
for i in {'1':1, (1,2):2, 3:5}:  # dict에서는 key만 반환하므로 for문에 사용 가능
    print(i, type(i))

print("==============")
    
for i in {'1':1, (1,2,):[2,2,], 3:5}.values():  
    print(i, type(i))

print("==============")
    
for i, j in {'1':1, (1,2,):[2,2,], 3:5}.items():  # unpacking 할당 방법
    print(i, type(i), '-', j, type(j))
print("==============")
for i, i in {'1':1, (1,2,):[2,2,], 3:5}.items():  # 같은 name을 사용하는 경우에 i가 key  할당 후 value가 할당 되기 때문에 for문 아래에서는 이미 i에 value가 할당된 상태 임
    print(i, type(i), '-', i, type(i))
    
    
```
