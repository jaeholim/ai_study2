```python
#class A(object):                       ## Object   ( )
class A:
    x = 1
    def __init__(self, a):     ## instance   
        print('init', a)       ## overriding :     __init__() ,   Method overriding 
        
a = A(1)
```

```python
class B:
    b = 2                   # class , class attribute : method   
    def __init__(self, a):  # (instalce) method : 
        self.a = a          # instance , instance attribute : method  
        
a = B(1)
b = B(2)
dir(a)
print(a.a, b.a)
print(a.b, b.b)
a.b = 5                      # instance attribute  "="      instance attribute 
                             #   class attribute 'b'  'a' instance  instance attribute 'b' 
                             #   : a self   self.b    instance  b  
        
print(a.b, b.b)              # b.b  instance   class attribute 
B.b = 10                     # class  class attribute 'b' 10 
                             # class attribute class name  
print(a.b, b.b)
a.b = 3
a.c = 20                     # instance attribute  "="      instance attribute 

print(a.b, b.b, a.c)
```

```python
class B:
    b = 2
    def __init__(self, a):
        b = a                 # instance method instance attribute  , class  class attribute  
                              #  code 'b = a' b __init__ method  local 
        
a = B(1)
print(a.b, B.b)
```

```python
class B:
    b = 2
    def __init__(self, a):
        self.a = a                 # instance method instance attribute  , class  class attribute  
                                   #  code 'b = a' b __init__ method  local 
    def jjj(self, c):              # jjj  instance attribute "c" instance  
        self.c = c
a = B(1)
print(a.b, B.b)
```
