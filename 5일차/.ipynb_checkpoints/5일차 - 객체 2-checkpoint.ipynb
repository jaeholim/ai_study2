# 객체
- 상속을 통해서 기능을 추가 삭제 할 수 있다
- 기본적으로 object를 상속 받고 있으며, object의 상속은 일반적으로 생략됨
- class의 method는 overriding 해서 재정의 할 수 있음

## class 변수와 instance 변수간의 메모리 구조


## class변수와 같은 이름의 instance 변수 생성 메모리 구조


- instance attribute에 있으면 instance attribute를 반환, 없으면 class attribute에서 찾아 반환
- instance attribute가 없을때 "="를 이용하여 값을 할당 하면 새로운 instance attribute가 생김


## __init__ 메서드의 local 변수를 class 변수인줄 알고 할 당할때 구조




```python
#class A(object):                       ## Object를 상속 받아 옴(생략되어 있음)
class A:
    x = 1
    def __init__(self, a):     ## instance화 될때 호출되는 함수
        print('init', a)       ## overriding : 이미 정이 되어 있는 __init__()를 재정의함, 객체안에 있는 Method만 overriding이 가능함
        
a = A(1)
```

```python
class B:
    b = 2                   # class 변수, class attribute : method 밖에 정의된 변수
    def __init__(self, a):  # (instalce) method : 
        self.a = a          # instance 변수, instance attribute : method내에 정의된 변수
        
a = B(1)
b = B(2)
dir(a)
print(a.a, b.a)
print(a.b, b.b)
a.b = 5                      # instance attribute가 없을때 "="를 이용하여 값을 할당 하면 새로운 instance attribute가 생김
                             # 현재 예제에서는 class attribute 'b'와는 다른 'a' instance에 대한 instance attribute 'b'가 생긴것임
                             # 추가 설명 : a가 self와 같기 때문에 self.b를 선언한것과 같게 되어 instance 변수 b가 추가로 생김
        
print(a.b, b.b)              # b.b의 경우 instance변수가 없기 때문에 class attribute에서 찾음
B.b = 10                     # class에 있는 class attribute 'b'를 10으로 변경
                             # class attribute는 class name으로 접근이 가능
print(a.b, b.b)
a.b = 3
a.c = 20                     # instance attribute가 없을때 "="를 이용하여 값을 할당 하면 새로운 instance attribute가 생김

print(a.b, b.b, a.c)
```

```python
class B:
    b = 2
    def __init__(self, a):
        b = a                 # instance method에서는 instance attribute만 바꿀수 있고, class 는 class attribute만 바꿀수 있음
                              # 해당 code에서 'b = a'의 b는 __init__ method에서만 사용되는 local변수 임
        
a = B(1)
print(a.b, B.b)
```

```python
class B:
    b = 2
    def __init__(self, a):
        self.a = a                 # instance method에서는 instance attribute만 바꿀수 있고, class 는 class attribute만 바꿀수 있음
                                   # 해당 code에서 'b = a'의 b는 __init__ method에서만 사용되는 local변수 임
    def jjj(self, c):              # jjj가 호출될때 instance attribute "c"가 instance 메모리에 생김
        self.c = c
a = B(1)
print(a.b, B.b)
```
