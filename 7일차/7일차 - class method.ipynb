# Class Method
- class method를 만들기 위해서는 "@classmethod" 데코레이터를 써줘야 함
- class attribute를 access 가능한 method이고
- class method는 class명으로 호출가능할 뿐 아니라 instance이름으로도 호출 가능


# __init__ 사용법


class A:
    def __init__(self):
        pass
       
a = A()           # method로 사용 : instance화를 통한 __init__() 호출
a.__init__()     # method로 사용 : 인스턴스의 Method 형태로 __init__() 호출
A.__init__(a)    # function으로 사용 : Class 이름을 이용하여 __init__ 호출 시에는 Function으로 사용

```python

class A:
    a = 1
    
    @classmethod   # class method : class 이름으로 접근 해야 하고, class attribute를 변경 할 수 있음
    def ss(cls):
        cls.a = 7
        
aa = A()
aa.ss()        # instanace도 class method를 호출 할 수 있고, 해당 class method에서 class attribute를 변경함
aa.a           # instance 로 'a' attribute 호출시 instance attribute에 a가 없기때문에 class attribute중 a를 찾음
```

```python
A.ss()
A.a
```

```python
a = A()
a.ss()
```
