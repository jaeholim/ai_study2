## delegation(위임) 방법
- 상속
- composition



```python
class A:
    def x(self):
        print('A')
class B(A):
    pass

b = B()
print(b.x())
print(B.x is A.x)    # class x와 상속 받은 class의 x는 부모의 x가리기는 것으로 둘은 같음
print(b.x is A.x)    # instance가 사용할때는 method, Class가 사용할때는 function으로 보기 때문에 instacne의 x와 class의 x는 다름

print(B.__mro__)     # Class간의 호출되는 순서
```

class x와 상속 받은 class의 x는 부모의 x가리기는 것으로 둘은 같음


```python
class A:
    def x(self):
        print('A')
class B(A):
    def x(self):
        print('A')
print(A.x is B.x)
```
