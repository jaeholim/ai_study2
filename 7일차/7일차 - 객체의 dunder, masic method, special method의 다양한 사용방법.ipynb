```python
class A:
    def __init__(self):
        self.a = 1
    def ma(self):
        pass

ai = A()
print(ai.__dict__['a'])
```

```python
try:
    ai.__dict__['b']
except KeyError as e:
    print('empty key : ', e)
else:
    print('try complete')
finally:
    print('complete')
    
print(ai.__class__().__dict__['a'])  # ai.__class__() --> Class A에 대한 instance (ai와는 다른 instance) --> 아래에 그림 설명 있음
print(ai.__class__.__dict__['__init__'])   # ai.__class__.__dict__ 는 mappingproxy dict이므로 키가 '__init__'를 가져 올 수 있음
print(ai.__class__.__dict__['ma'])   # ai.__class__.__dict__ 는 mappingproxy dict이므로 키가 '__init__'를 가져 올 수 있음
print(ai.__class__.__dict__['a'])   # ai.__class__ --> Class A이므로 class 변수를 접근
```
